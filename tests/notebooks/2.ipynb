### Filtering

```python
df['Year'] = df['date'].apply(lambda x: x[:4])
df = pd.DataFrame(df[df['Year'] == '2016'])
df.drop(['Year'], axis = 1, inplace = True)
df
```

# Date Conversion

```python
converter = DateConverter()
df['date'] = df['date'].apply(lambda x: converter.convert_date(x, infer_format=True))
df
```

# Models

## ARIMA

```python
from tsa.models import Arima
```

```python
train_df = df[:int(0.8 * len(df))]
test_df = df[int(0.8 * len(df)):]

df = train_df
```

```python
model = Arima()
model(df, 'date', 'wind_speed')
```

```python
model.test_stationarity()
```

```python
params = model.acf_plot(0)
```

```python
params.keys()
```

```python
outfile(params['title'])
```

```python
params = model.pacf_plot(0)
```

```python
outfile(params['lower'])
```

```python
model.fit()
```

```python
model.error_metrics(test_df, 'wind_speed')
```

```python
params = model.plot_forecast(test_df, 'wind_speed', exclude_time=True)
```

```python
outfile(params['y_forecast'])
```
