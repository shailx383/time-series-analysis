### Filtering

```python
def filter_year(datetime: str):
    slash = 0
    for i in range(len(datetime)):
        if datetime[i] == '/':
            slash += 1
        if slash == 2:
            return str(datetime[i+1:i+5])
    
df['Year'] = df['Date'].apply(lambda x: filter_year(x))
df        
```

```python
df = pd.DataFrame(df.iloc[::-1])
df = df.reset_index().drop('index', axis = 1)
df
```

```python
df = pd.DataFrame(df[df['Year'] == '2021'])
```

# Date Conversion

```python
converter = DateConverter()
converter.directive_df()
```

```python
df['Date'] = df['Date'].apply(lambda x: converter.convert_date(x, include_time=True, format_str="%m/%d/%Y %H:%M"))
df
```
