```python
import tsa.dateconverter as converter
import pandas as pd
```

```python
df= pd.read_csv('/content/Rainfall_data.csv')
```

```python
df
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df['Time'] = df.apply(lambda x: str(int(x['Year']))+'-'+str(int(x['Month']))+'-'+str(int(x['Day'])),axis = 1)
```

```python
df['Time']
```

```python
formatter = converter.DateConverter()
```

```python
formatter.directive_df()
```

```python
df['Time'] = df['Time'].apply(lambda x: formatter.convert_date(x))
```

```python
df['Time']
```

```python
df['Time'] = pd.to_datetime(df['Time'])
```
